As Sport and Male variables are categorical. To determine the relationship between these 2 variables we need chi-square test

```python
from scipy.stats import chi2_contingency
chiVal, pVal, degree, exp = chi2_contingency(cross_tab1)
chiVal, pVal, degree, exp
```

The first value is the chi-square value of 57.29. The chance of such a value or even more extreme, in a sample, if there is no association in the population is 2.22*10-12 (the second value). Therefore the p-value significance value is below 0.05. In this case it indicates that there's an association between the two variables

## 12.1 -  There is an asscociation between sport and gender <br> <br>
## 12.2 - The test statistic is calculated by the chi2_contingency function = 57.29 <br> <br>
## 12.3 - The p-value will also be calculated by the chi2_contingency function - 1.5768739806720535e-09 <br> <br>
## 12.4 - We compare the p-value with the significance level of 0.05 (to determine whether the association is statistically significant). Therefore the association is statistically significant <br> <br>
## 12.5 - The assumptions of the chi-square test for independence include:
## Independence: The observations in the sample should be independent of each other. This means that each observation should not be influenced by any other observation in the sample.
## Sample size: The sample size should be large enough to ensure that the expected count in each cell of the contingency table is at least 5.
## Random sampling: The sample should be selected using a random sampling method to ensure that it is representative of the population.
## Measurement scale: The variables being tested should be measured at a nominal or ordinal scale.
